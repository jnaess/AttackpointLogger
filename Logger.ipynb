```python
from helium import *
import random
import time
import threading

def delay(min = 2, max = 5):
    """
        Pauses program from 3 to 15 seconds. Randomly picks a number 
        in between 3 and 15 to the tenth of a millisecond
    """
    sleeptime = random.randrange(min*1000,max*1000,step=1)
    time.sleep(sleeptime/1000)
    
class element():
    def __init__(self, Xmin, Ymin, Xmax, Ymax, string=''):
        self.Xmin = Xmin
        self.Ymin = Ymin
        self.Xmax = Xmax
        self.Ymax = Ymax
        self.string = string
        
    def random_location(self):
        """
        example call: random_location(1,1,50,50)
        Function randomized where the clicks location to simulate human inconsistency
        """
        self.X = random.randrange(self.Xmin,self.Xmax,step=1)
        self.Y = random.randrange(self.Ymin,self.Ymax,step=1)

    def clicker(self):
        """
        Delays before click
        Clicks the specified button location
        """
        delay()
        self.random_location()
        click(Point(self.X,self.Y))
        
    def do(self):
        """
        Delays before entry
        Filled the entry with the elements string
        """
        self.clicker()
        if(self.string != ''):
            delay()
            write(self.string)
        
class city_scroll():
    def __init__(self, cities_in_scroll = []):
        self.cities_in_scroll = cities_in_scroll
        self.confirm = element(743,470,822,489)
        self.accept_loss = element(391,467,505,486)
        
    def iter_cities(self):
        """
        Iterate through each city within this scroll view and demand their respective resources
        """
        for city in self.cities_in_scroll:
            city.do()
            self.confirm.do()
            self.accept_loss.do()
        
class browser():
    def __init__(self, city_list = []):
        self.city_list = city_list
        
        #login info
        self.email_box = element(518,38,650,58, 'super.ppl.1k@gmail.com')
        self.pass_box = element(666,37,800,62,'Prospect90')
        self.login_button = element(810,36,868,63)
        self.world_button = element(580,296,727,327)
        self.captain = True
                
        #display pages
        self.island_view = element(35,6,64,35)
        self.city_view = element(72,7,132,33)
        self.city_jump = element(104,24,132,53)
        
        #switch city
        self.switch_city = element(592,4,603,22)
        self.scroll_up = element(332,115,342,120)
        self.scroll_down = element(332,465,342,478)
        
        #farming village commands
        self.farming_exit= element(866,30,883,46)
        self.demand_five = element(186,351,318,510)
        self.captain = element(194,2,219,27)
        self.confirm = element(743,470,822,489)
        self.accept_loss = element(391,467,505,486)
        self.captain_exit = element(883,32,901,48)
        self.captin_select = element(201,310,339,324)
        self.delete_reports = element(859,472,872,487)
        
    def login(self):
        """
        Login to the city
        """
        self.email_box.do()
        self.pass_box.do()
        self.login_button.do()
        self.world_button.do()
        
    def demand(self, counter, time=5, duration = 0):
        """
        Demand for a farming village
        Iterate through each scroll view of cities and then move to next view
        """
        print("Island Demanded")
                
        if counter%250 == 0:
            for cities in self.city_list:
                cities.iter_cities()
                self.scroll_Down()
            self.scroll_Up(8)
            
    def scroll_Up(self, times=1):
        count = 0
        while count < times:
            self.scroll_up.do()
            count = count + 1
            
    def scroll_Down(self, times=1):
        count = 0
        while count < times:
            self.scroll_down.do()
            count = count + 1
            

        
scroll_zero = [element(132,140,342,178),element(130,232,343,266),
              element(130,324,343,354),element(130,416,343,440),
              element(239,471,326,496)]
view_a = city_scroll(scroll_zero)

scroll_one = [element(132,221,329,258),element(131,308,328,345),
              element(131,394,327,431),element(132,482,328,494)]
view_b = city_scroll(scroll_one)

scroll_two = [element(131,215,327,251),element(131,302,329,354),
              element(130,405,328,440),element(201,472,327,496)]
view_c = city_scroll(scroll_two)

scroll_three = [element(132,224,329,261),element(131,331,328,348),
              element(130,399,328,436),element(254,466,328,494)]
view_d = city_scroll(scroll_three)

scroll_last = [element(123,94,326,130),element(131,182,343,212),
              element(130,268,323,298),element(131,354,343,384),
              element(131,440,343,470)]
view_e = city_scroll(scroll_last)

city_list = [view_a,view_b,view_c,view_d,view_a,view_a,view_a]
        
        
grepolis = browser(city_list)

driver = start_chrome('https://en.grepolis.com/')
set_driver(driver)

#Start up grapolis and do first round of remanding resources
grepolis.login()

counter = 0
continue_loop = True

while continue_loop:
    #check for demands being ready
    #intervals of 615 seconds
    grepolis.demand(counter)
    
    #run program every 1 seconds
    time.sleep(1)
    
    counter = counter + 1
    print(counter)
    #Check to break loop
    #Insert integer number of demands you want to run
    break_point = 1000
    if counter / 545 > break_point:
        continue_loop = False



#kill_browser()
```
